```python
## using list comprehension to find the features with null values
features_with_null_values = [features for features in df.columns if df[features].isnull().sum()  > 1]

## using for loop to print the percentage of the null values present in the dataset
for features in features_with_null_values:
    print(f"{features} has : {np.round(df[features].isnull().mean(),4)} % of missing values")
```

```python
## priting the total number of the features with missing values
print(len(features_with_null_values))
```

### since we have 18 missing values, first we need to understand the relationship between those features with the salePrice of the houses

```python
for feature in features_with_null_values:
    data = df.copy()
    # let's make a variable that indicates 1 if the observation was missing or zero otherwise
    data[feature] = np.where(data[feature].isnull(), 1, 0)
    # let's calculate the mean SalePrice where the information is missing or present
    null_fet = data.groupby(feature)['saleprice'].median()
    sns.barplot(null_fet, palette="Set1")
    plt.title("A graph between"+ " " + feature + " " + "null vlaues vs not null values")
    plt.show()
```

#### now seperating the numerical vs categorical columns of the data


```python
## seperating the numerical features of the dataframe
numerical_features = df.select_dtypes(include=(np.number)).isin(["0.0","1.0"]).astype(bool).columns.tolist()
print(f"The total number of numerical features: {len(numerical_features)}")
```

```python
### seperating the categorical columns of the dataframe
categorical_features = df.select_dtypes(include =["object"]).columns
print(f"The total number of catrgorical features: {len(categorical_features)}")
```

```python
## printing the features that have year in the dataframe
year_feature = [feature for feature in numerical_features if 'yr' in feature or 'year' in feature]
print(year_feature)
```

```python
for YEARS in year_feature:
    print(YEARS, df[YEARS].unique())
```

```python
### lets see is there any relationship between yrsold and price of the houses
yrSold_price = df.groupby("yrsold")['saleprice'].median()
sns.lineplot(data=yrSold_price)
plt.title("Years vs Median SalePrice of the houses")
plt.plot()
```

```python
for yearfeature in year_feature:
    if yearfeature != "yrsold":
        df = df.copy()
        df[yearfeature] = df['yrsold'] - df[yearfeature]
        sns.scatterplot(x=df[yearfeature],y=df["saleprice"])
        plt.xlabel(yearfeature)
        plt.ylabel("SalesPrice")
        plt.show()
```

```python
## finding the descrete feature present in the numerical feature columns

discrete_feature = [feature for feature in numerical_features if len(df[feature].unique())<25]
print(f'Discrete Variables count: {len(discrete_feature)}')
```

```python
discrete_feature
```

```python
## printing the head of the descrete feature
df[discrete_feature].head()
```

```python
### lets find the relationship between discreate feature and sale price

for feature in discrete_feature:
    df = df.copy()
    desc = df.groupby(feature)["saleprice"].median()
    sns.barplot(desc, palette="Set2")
    plt.show()
```

```python
## continuous variable from the dataset
continuous_feature = [feature for feature in numerical_features if feature not in discrete_feature + year_feature]
print(f"Continuous feature Count {len(continuous_feature)}")
```
