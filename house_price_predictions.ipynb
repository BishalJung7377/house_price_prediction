```python
## using list comprehension to find the features with null values
features_with_null_values = [features for features in df.columns if df[features].isnull().sum()  > 1]

## using for loop to print the percentage of the null values present in the dataset
for features in features_with_null_values:
    print(f"{features} has : {np.round(df[features].isnull().mean(),4)} % of missing values")
```

```python
## priting the total number of the features with missing values
print(len(features_with_null_values))
```
